### Questão 1.2

Considere o cógido abaixo para gerar os dados. Usando o algoritmo k-means e a medida Normalized Mutual Information (NMI), determine o número ideal de clusters. Dica: use a função normalized_mutual_info_score do sklearn.

<font color='red'>a.4</font><br>
b.8<br>
c.6<br>
d.5<br>
e.2<br>

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
np.random.seed(201)

import numpy as np
import matplotlib.pyplot as plt
# create blobs
c = [(-0,1),(0,1),(5,6),(3,1),(6,9)]
n=300
data = make_blobs(n_samples=n, n_features=2, centers=c, cluster_std=1, random_state=50)
X = data[0]
labels = data[1]
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=50, alpha=0.9)
plt.show(True)
```

```python
# Buscando o k ótimo
from scipy.spatial.distance import cdist
from sklearn.metrics.cluster import normalized_mutual_info_score

NMI_km = []
K = range(1,10)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    y_km = kmeans.fit_predict(X)
    NMI_km.append(normalized_mutual_info_score(y_km, data[1], average_method='arithmetic'))

plt.figure(figsize=(10,5))
plt.plot(K, NMI_km, '-bo', label = 'kmeans')
plt.xlabel('k', fontsize = 15)
plt.ylabel('Normalized Mutual information', fontsize = 15)
plt.grid(True)
plt.legend()
plt.show()

best_k = 0
best_NMI = 0

for k, NMI in enumerate(NMI_km, start=1):
    if NMI > best_NMI:
        best_NMI = NMI
        best_k = k
print('best_k= ', best_k, '\nbest_NMI= ', best_NMI)
```
